```python
import requests
from bs4 import BeautifulSoup
```

```python
mylink = 'https://jiji.co.ke/cars?filter_attr_1_make=Toyota&filter_attr_2_model=Land%20Cruiser'
```

```python
mylink
```

```python
response = requests.get(mylink)
result = BeautifulSoup(response.content, "html.parser")
```

```python
result
```

```python
Carnames = result.find_all(class_='b-list-advert-base qa-advert-list-item b-list-advert-base--gallery')
```

```python
Carnames
```

```python
links = [l.get('href') for l in Carnames]
```

```python
links
```

```python
links = [f'http://jiji.co.ke{l}' for l in links]
```

```python
links
```

```python


scraped_titles = []

for link in links:
    response = requests.get(link)
    myresult = BeautifulSoup(response.content, "html.parser")

   
    title_element = myresult.find(class_="b-advert-title-inner qa-advert-title b-advert-title-inner--h1")

    
    if title_element:
        title = title_element.get_text(strip = True)
        scraped_titles.append(title)
    else:
        scraped_titles.append("Title not found for this link")


print(scraped_titles)
```

```python
import pandas as pd
```

```python
mydataframe = pd.DataFrame({"Product Name" : scraped_titles })
```

```python
mydataframe
```

```python


scraped_prices = []

for link in links:
    response = requests.get(link)
    myresult = BeautifulSoup(response.content, "html.parser")

   
    title_element = myresult.find(class_="qa-advert-price-view b-alt-advert-price__container")

    
    if title_element:
        title = title_element.get_text(strip = True)
        scraped_prices.append(title)
    else:
        scraped_titles.append("Title not found for this link")


print(scraped_prices)
```

```python
mydataframe = pd.DataFrame({"Product Name" : scraped_titles , "Prices" : scraped_prices  })
```

```python
mydataframe
```

```python


scraped_modal = []

for link in links:
    response = requests.get(link)
    myresult = BeautifulSoup(response.content, "html.parser")

   
    title_element = myresult.find(attrs={"itemprop": "model"})
    title_element = myresult.find(attrs={"aria-label": "bathrooms"})

    
    if title_element:
        title = title_element.get_text(strip = True)
        scraped_modal.append(title)
    else:
        scraped_modal.append("Title not found for this link")


print(scraped_modal)
```

```python


scraped_color = []

for link in links:
    response = requests.get(link)
    myresult = BeautifulSoup(response.content, "html.parser")

   
    title_element = myresult.find(attrs={"itemprop": "color"})

    
    if title_element:
        title = title_element.get_text(strip = True)
        scraped_color.append(title)
    else:
        scraped_color.append("Title not found for this link")


print(scraped_color)
```

```python
mydataframe = pd.DataFrame({"Product Name" : scraped_titles , "Prices" : scraped_prices , "Modal" : scraped_modal , "color" : scraped_color  })
```
